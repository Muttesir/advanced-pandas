```python
df_a_s=df[['Age','Sex']].copy()
df_a_s.head(5)
```

```python
def get_gennder_bool(gender):
    
    if gender =='male':
        return 1
    else:
        return 0
    
```

```python
df_a_s['bool_male']=df_a_s.Sex.apply(get_gennder_bool)
df_a_s
```

```python
def get_older(age):
    if age >= 60:
        return 'retired'
    else:
        return 'still working'
```

```python
df_a_s['older']=df_a_s.Age.apply(get_older)
df_a_s.head(20)
```

```python
df_name=df[['Name', 'Age']].copy()
df_name
```

```python
def get_title(name):
    parsed_name=name.split(" ")
    title=parsed_name[1]
    return title
```

```python
df_name['Title']=df_name.Name.apply(get_title)
df_name.head()
```

```python
df_a_s=df[['Age','Sex']].copy()
df_a_s.head(5)
```

```python
def old_man(row):
    age=row['Age']
    sex=row['Sex']
    if age >=60 and sex == 'male':
        return 1
    else:
        return 0
```

```python
df_a_s['Old_man']=df_a_s.apply(old_man, axis=1)
df_a_s['Old_man'].sum()
```
